# Series de Taylor

```python
import numpy as np
import matplotlib.pyplot as plt

# Define el rango de los valores de x
x = np.linspace(-5,5 ,400)

# Calcula la funcin coseno
cos_y = np.cos(x)

# Calcula la expansin de Taylor aproximada al conseno
taylor_y = 1 - (x**2)/2 + (x**4)/24 - (x**6)/720

# Crea la grfica
plt.figure(figsize=(10, 6))
plt.plot(x, cos_y, label='cos(x)')
plt.plot(x, taylor_y, label='Expansin de Taylor')

# Agrega etiquetas y titulo
plt.xlabel('x')
plt.ylabel('y')
plt.title('Funcin de Coseno y la Expansin de Taylor')
plt.grid(True)
plt.legend()
plt.show()
```
