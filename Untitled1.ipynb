#**PARABOLA**

```python
import matplotlib.pyplot as plt
import numpy as np

x = [i for i in range(-10, 11)]
y = [pow(i,2) for i in x]

plt.plot(x,y)
plt.show()
```

Este código realiza la gráfica de una parabola  $$y = x^2$$
